```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate Random Data (Simulating a Dataset)
n_samples = 500
n_features = 10
np.random.seed(42)
data = np.random.rand(n_samples, n_features)  # Random values between 0 and 1
labels = np.random.choice([0, 1], size=n_samples)  # Binary labels for visualization
```

```python
# Step 2: Preprocessing (Feature Scaling)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(data_scaled, labels, test_size=0.2, random_state=42)
```

```python
# Step 4: Train Base Model (t-SNE)
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_train_tsne = tsne.fit_transform(X_train)
```

```python
# Step 5: Evaluate (Visualization)
plt.figure(figsize=(8, 6))
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1], c=y_train, cmap='viridis', alpha=0.7)
plt.colorbar(label='Label')
plt.title('t-SNE Visualization of Training Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

```python
# Step 6: Predict New Data (Applying t-SNE on Test Data)
X_test_tsne = tsne.fit_transform(X_test)

# Visualizing Test Data
plt.figure(figsize=(8, 6))
plt.scatter(X_test_tsne[:, 0], X_test_tsne[:, 1], c=y_test, cmap='plasma', alpha=0.7)
plt.colorbar(label='Label')
plt.title('t-SNE Visualization of Test Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
