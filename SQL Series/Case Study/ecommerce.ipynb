```python
from sqlalchemy import create_engine, text
```

```python
# Database connection details
username = "xxxxxx"
password = "xxxxxx"
host = "localhost"
port = "3306"
database = "ecommerce"

# Create the engine
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
```

```python
# Connect to the database and execute a query to list all tables
try:
    with engine.connect() as connection:
        result = connection.execute(text("SHOW TABLES;"))
        tables = [row[0] for row in result]
        print("Tables in the 'ecommerce' database:")
        for table in tables:
            print(table)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd
from IPython.display import display

query = "SELECT * FROM customers;"
customer_df = pd.read_sql(query, engine)
display(customer_df)
```

```python
df = pd.read_sql("SELECT customer_id, name, email FROM customers;", engine)
display(df)
```

```python
df = pd.read_sql("SELECT * FROM customers ORDER BY created_at DESC;", engine)
display(df)
```

```python
df = pd.read_sql("SELECT COUNT(*) AS total_customers FROM customers;", engine)
display(df)
```

```python
df = pd.read_sql("SELECT address, COUNT(*) AS customer_count FROM customers GROUP BY address;", engine)
display(df)
```
