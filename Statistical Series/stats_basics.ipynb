```python
# Sample Data
data = np.random.normal(loc=50, scale=15, size=100)  # Generating random data

# a. Range
range_value = np.ptp(data)
print("Range:", range_value)

# b. Mean Absolute Deviation
mad = np.mean(np.abs(data - np.mean(data)))
print("Mean Absolute Deviation:", mad)

# c. Variance
variance = np.var(data, ddof=1)
print("Variance:", variance)

# d. Standard Deviation
std_dev = np.std(data, ddof=1)
print("Standard Deviation:", std_dev)

# e. Coefficient of Variation
cv = std_dev / np.mean(data)
print("Coefficient of Variation:", cv)

# f. IQR
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
print("Interquartile Range (IQR):", iqr)

# g. Skewness & Kurtosis
skewness = skew(data)
kurt = kurtosis(data)
print("Skewness:", skewness)
print("Kurtosis:", kurt)

# h. Outliers (using IQR method)
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers:", outliers)

# i. Median Absolute Deviation
mad_stat = median_abs_deviation(data)
print("Median Absolute Deviation:", mad_stat)

# j. Variance to Mean Ratio
vmr = variance / np.mean(data)
print("Variance to Mean Ratio:", vmr)

# k. Covariance & Correlation
# Generating a second dataset
data2 = np.random.normal(loc=55, scale=10, size=100)
dataframe = pd.DataFrame({'Variable1': data, 'Variable2': data2})
covariance = dataframe.cov()
correlation = dataframe.corr()
print("Covariance Matrix:\n", covariance)
print("Correlation Matrix:\n", correlation)

# Visualization for Correlation Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Probability
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde, skew, kurtosis, median_abs_deviation
from itertools import permutations, combinations
```

```python
# Sample Data
data = np.random.normal(loc=50, scale=15, size=100)  # Generating random data

# a. Probability of an Event
def probability_of_event(successes, total_outcomes):
    return successes / total_outcomes

print("Probability of an Event (Rolling a 6 on a fair die):", probability_of_event(1, 6))

# b. Sample Spaces
sample_space_dice = list(range(1, 7))  # Rolling a die
print("Sample Space (Die Roll):", sample_space_dice)

# c. Events
# 1) Mutually Exclusive Events
mutually_exclusive = set([1, 2, 3]).isdisjoint(set([4, 5, 6]))
print("Are events {1,2,3} and {4,5,6} mutually exclusive?", mutually_exclusive)

# 2) Independent Events
def independent_probability(p_a, p_b):
    return p_a * p_b

print("Independent Probability (Rolling two fair dice and getting two 6s):", independent_probability(1/6, 1/6))

# d. Combinations & Permutations
def compute_permutations(n, r):
    return len(list(permutations(range(n), r)))

def compute_combinations(n, r):
    return len(list(combinations(range(n), r)))

print("Permutations (P(5,3)):", compute_permutations(5, 3))
print("Combinations (C(5,3)):", compute_combinations(5, 3))

# e. Probability of Multiple Events
# 1) Joint Probability (P(A and B) = P(A) * P(B) for independent events)
print("Joint Probability (Getting a 6 on two dice rolls):", independent_probability(1/6, 1/6))

# 2) Union Probability (P(A or B) = P(A) + P(B) - P(A and B))
def union_probability(p_a, p_b, p_a_and_b):
    return p_a + p_b - p_a_and_b

print("Union Probability (Rolling a 1 or a 6 on one die):", union_probability(1/6, 1/6, 0))

# 3) Marginal Probability (Sum of Joint Probabilities)
def marginal_probability(joint_probs):
    return sum(joint_probs)

print("Marginal Probability Example:", marginal_probability([0.1, 0.2, 0.15]))

# f. Event Dependencies
# 1) Independent Probability (P(A and B) = P(A) * P(B))
print("Independent Probability (Example):", independent_probability(0.4, 0.5))

# 2) Dependent Probability (P(A|B) = P(A and B) / P(B))
def dependent_probability(p_a_and_b, p_b):
    return p_a_and_b / p_b

print("Dependent Probability (P(A|B) where P(A and B) = 0.2 and P(B) = 0.5):", dependent_probability(0.2, 0.5))

# 3) Conditional Independence (P(A and B | C) = P(A | C) * P(B | C))
def conditional_independence(p_a_given_c, p_b_given_c):
    return p_a_given_c * p_b_given_c

print("Conditional Independence Example:", conditional_independence(0.3, 0.4))
```

```python
# g. Types of Probability
classical_prob = probability_of_event(1, 6)
empirical_prob = 50 / 100  # Example of observed frequency
subjective_prob = 0.7  # Example of expert opinion
print("Classical Probability:", classical_prob)
print("Empirical Probability:", empirical_prob)
print("Subjective Probability:", subjective_prob)

# h. Bayesian vs. Frequentist Interpretation
# Bayesian updates probability based on evidence, Frequentist relies on long-term frequency

# i. Probability Axioms
# 1) Non-negativity: P(A) >= 0
# 2) Normalization: P(Sample Space) = 1
# 3) Additivity: P(A or B) = P(A) + P(B) for disjoint events

# j. Extended Additivity
# If A1, A2, ... are disjoint, then P(Union Ai) = Sum P(Ai)

# k. Probability Rules
# 1) Complement Rule: P(A') = 1 - P(A)
def complement_rule(p_a):
    return 1 - p_a

print("Complement Rule (P(not A) if P(A) = 0.3):", complement_rule(0.3))

# 2) Addition Rule: P(A or B) = P(A) + P(B) - P(A and B)
print("Addition Rule (P(A=0.4, B=0.5, A and B=0.2)):", union_probability(0.4, 0.5, 0.2))

# 3) Multiplication Rule: P(A and B) = P(A|B) * P(B)
print("Multiplication Rule (P(A|B) = 0.6, P(B) = 0.5):", dependent_probability(0.3, 0.5))

# l. Law of Total Probability
# P(A) = Sum P(A|Bi)P(Bi) for partitions Bi of sample space

# m. Conditional Probability: P(A|B) = P(A and B) / P(B)
print("Conditional Probability (Example):", dependent_probability(0.2, 0.5))

# n. Bayes' Theorem: P(A|B) = [P(B|A) * P(A)] / P(B)
def bayes_theorem(p_b_given_a, p_a, p_b):
    return (p_b_given_a * p_a) / p_b

print("Bayes' Theorem (P(B|A)=0.7, P(A)=0.4, P(B)=0.5):", bayes_theorem(0.7, 0.4, 0.5))

# 1) Bayesian Inference (Priors & Posteriors)
prior = 0.4
likelihood = 0.7
evidence = 0.5
posterior = bayes_theorem(likelihood, prior, evidence)
print("Posterior Probability:", posterior)
```

```python
# Probability Distribution
# Bernoulli Distribution
from scipy.stats import bernoulli
```

```python
# Bernoulli Distribution
p = 0.3  # Probability of success

# Generate Bernoulli random variables
bernoulli_samples = bernoulli.rvs(p, size=1000)

# Expected Value & Variance
expected_value = p
variance = p * (1 - p)
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

# Probability Mass Function (PMF)
k = [0, 1]
pmf_values = bernoulli.pmf(k, p)
plt.bar(k, pmf_values, tick_label=k)
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Bernoulli PMF")
plt.show()

# Cumulative Distribution Function (CDF)
cdf_values = bernoulli.cdf(k, p)
plt.step(k, cdf_values, where="post")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.title("Bernoulli CDF")
plt.show()

# Moment-Generating Function (MGF)
def bernoulli_mgf(t, p):
    return (1 - p) + p * np.exp(t)

t_values = np.linspace(-2, 2, 100)
mgf_values = bernoulli_mgf(t_values, p)
plt.plot(t_values, mgf_values)
plt.xlabel("t")
plt.ylabel("MGF")
plt.title("Bernoulli Moment-Generating Function")
plt.show()

# Survival Function (1 - CDF)
survival_values = 1 - bernoulli.cdf(k, p)
survival_values[survival_values == 0] = np.nan  # Avoid division by zero
plt.step(k, survival_values, where="post")
plt.xlabel("Outcome")
plt.ylabel("Survival Probability")
plt.title("Bernoulli Survival Function")
plt.show()

# Hazard Function (PMF / Survival Function)
hazard_values = np.divide(pmf_values, survival_values, out=np.zeros_like(pmf_values), where=survival_values!=0)
plt.step(k, hazard_values, where="post")
plt.xlabel("Outcome")
plt.ylabel("Hazard Rate")
plt.title("Bernoulli Hazard Function")
plt.show()
```

```python
from scipy.stats import binom

# Binomial Distribution
n = 10  # Number of trials
p = 0.3  # Probability of success per trial

# Generate Binomial random variables
binomial_samples = binom.rvs(n, p, size=1000)

# Expected Value & Variance
expected_value = n * p
variance = n * p * (1 - p)
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

# Probability Mass Function (PMF)
k = np.arange(0, n + 1)
pmf_values = binom.pmf(k, n, p)
plt.bar(k, pmf_values, tick_label=k)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial PMF")
plt.show()

# Cumulative Distribution Function (CDF)
cdf_values = binom.cdf(k, n, p)
plt.step(k, cdf_values, where="post")
plt.xlabel("Number of Successes")
plt.ylabel("Cumulative Probability")
plt.title("Binomial CDF")
plt.show()

# Moment-Generating Function (MGF)
def binomial_mgf(t, n, p):
    return (1 - p + p * np.exp(t)) ** n

t_values = np.linspace(-2, 2, 100)
mgf_values = binomial_mgf(t_values, n, p)
plt.plot(t_values, mgf_values)
plt.xlabel("t")
plt.ylabel("MGF")
plt.title("Binomial Moment-Generating Function")
plt.show()

# Survival Function (1 - CDF)
survival_values = 1 - binom.cdf(k, n, p)
survival_values[survival_values == 0] = np.nan  # Avoid division by zero
plt.step(k, survival_values, where="post")
plt.xlabel("Number of Successes")
plt.ylabel("Survival Probability")
plt.title("Binomial Survival Function")
plt.show()

# Hazard Function (PMF / Survival Function)
hazard_values = np.divide(pmf_values, survival_values, out=np.zeros_like(pmf_values), where=survival_values!=0)
plt.step(k, hazard_values, where="post")
plt.xlabel("Number of Successes")
plt.ylabel("Hazard Rate")
plt.title("Binomial Hazard Function")
plt.show()
```

```python
from scipy.stats import poisson

# Poisson Distribution
lambda_ = 4  # Average rate (mean number of occurrences in a fixed interval)

# Generate Poisson random variables
poisson_samples = poisson.rvs(lambda_, size=1000)

# Expected Value & Variance
expected_value = lambda_
variance = lambda_
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

# Probability Mass Function (PMF)
k = np.arange(0, 15)
pmf_values = poisson.pmf(k, lambda_)
plt.bar(k, pmf_values, tick_label=k)
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Poisson PMF")
plt.show()

# Cumulative Distribution Function (CDF)
cdf_values = poisson.cdf(k, lambda_)
plt.step(k, cdf_values, where="post")
plt.xlabel("Number of Events")
plt.ylabel("Cumulative Probability")
plt.title("Poisson CDF")
plt.show()

# Moment-Generating Function (MGF)
def poisson_mgf(t, lambda_):
    return np.exp(lambda_ * (np.exp(t) - 1))

t_values = np.linspace(-2, 2, 100)
mgf_values = poisson_mgf(t_values, lambda_)
plt.plot(t_values, mgf_values)
plt.xlabel("t")
plt.ylabel("MGF")
plt.title("Poisson Moment-Generating Function")
plt.show()

# Survival Function (1 - CDF)
survival_values = 1 - poisson.cdf(k, lambda_)
survival_values[survival_values == 0] = np.nan  # Avoid division by zero
plt.step(k, survival_values, where="post")
plt.xlabel("Number of Events")
plt.ylabel("Survival Probability")
plt.title("Poisson Survival Function")
plt.show()

# Hazard Function (PMF / Survival Function)
hazard_values = np.divide(pmf_values, survival_values, out=np.zeros_like(pmf_values), where=survival_values!=0)
plt.step(k, hazard_values, where="post")
plt.xlabel("Number of Events")
plt.ylabel("Hazard Rate")
plt.title("Poisson Hazard Function")
plt.show()

# Additional Exploration: Mean and Variance Verification
sample_mean = np.mean(poisson_samples)
sample_variance = np.var(poisson_samples)
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

```python
from scipy.stats import hypergeom, multinomial

# Categorical Distribution
p = [0.2, 0.3, 0.5]  # Probabilities for each category
n = 1000  # Number of trials

# Generate Categorical random variables
categorical_samples = np.random.choice(len(p), size=n, p=p)

# Expected Value & Variance
expected_values = np.array(p) * n
variances = np.array(p) * (1 - np.array(p)) * n
print(f"Expected Values: {expected_values}")
print(f"Variances: {variances}")

# Probability Mass Function (PMF)
k = np.arange(len(p))
k_counts = np.bincount(categorical_samples, minlength=len(p)) / n
plt.bar(k, k_counts, tick_label=[f"Category {i}" for i in k])
plt.xlabel("Category")
plt.ylabel("Probability")
plt.title("Categorical PMF (Estimated from Samples)")
plt.show()

# Cumulative Distribution Function (CDF)
cdf_values = np.cumsum(p)
plt.step(k, cdf_values, where="post")
plt.xlabel("Category")
plt.ylabel("Cumulative Probability")
plt.title("Categorical CDF")
plt.show()

# Survival Function (1 - CDF)
survival_values = 1 - cdf_values
plt.step(k, survival_values, where="post", linestyle="dashed", color="red")
plt.xlabel("Category")
plt.ylabel("Survival Probability")
plt.title("Categorical Survival Function")
plt.show()

# Hazard Function (PMF / Survival Function)
hazard_values = np.divide(k_counts, survival_values, out=np.zeros_like(k_counts), where=survival_values != 0)
plt.step(k, hazard_values, where="post", linestyle="dotted", color="green")
plt.xlabel("Category")
plt.ylabel("Hazard Rate")
plt.title("Categorical Hazard Function")
plt.show()

# Additional Exploration: Sample Mean and Variance Verification
sample_means = np.bincount(categorical_samples, minlength=len(p)) / n
sample_variances = sample_means * (1 - sample_means) * n
print(f"Sample Means: {sample_means}")
print(f"Sample Variances: {sample_variances}")
```

```python
# Continuous Probability Distribution
from scipy.stats import multinomial, uniform

# Uniform Distribution Parameters
a, b = 0, 10  # Lower and upper bounds

# Generate Uniform random variables
uniform_samples = uniform.rvs(a, b-a, size=1000)

# Expected Value & Variance
expected_value = (a + b) / 2
variance = ((b - a) ** 2) / 12
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

# Probability Density Function (PDF)
x = np.linspace(a, b, 100)
pdf_values = uniform.pdf(x, a, b-a)
plt.plot(x, pdf_values, label="PDF", color='blue')
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Uniform Distribution PDF")
plt.legend()
plt.show()

# Cumulative Distribution Function (CDF)
cdf_values = uniform.cdf(x, a, b-a)
plt.plot(x, cdf_values, label="CDF", color='orange')
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.title("Uniform Distribution CDF")
plt.legend()
plt.show()

# Survival Function (1 - CDF)
survival_values = 1 - cdf_values
plt.plot(x, survival_values, label="Survival Function", linestyle="dashed", color="red")
plt.xlabel("x")
plt.ylabel("Survival Probability")
plt.title("Uniform Distribution Survival Function")
plt.legend()
plt.show()

# Hazard Function (PDF / Survival Function)
hazard_values = np.divide(pdf_values, survival_values, out=np.zeros_like(pdf_values), where=survival_values != 0)
plt.plot(x, hazard_values, label="Hazard Function", linestyle="dotted", color="green")
plt.xlabel("x")
plt.ylabel("Hazard Rate")
plt.title("Uniform Distribution Hazard Function")
plt.legend()
plt.show()

# Additional Exploration: Sample Mean and Variance Verification
sample_mean = np.mean(uniform_samples)
sample_variance = np.var(uniform_samples)
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

```python
from scipy.stats import expon

# Parameters for Exponential Distribution
lambda_ = 1.5  # Rate parameter

# Expected Value & Variance
expected_value = 1 / lambda_
variance = 1 / lambda_**2
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

# Generate x values
x = np.linspace(0, 5, 1000)

# Probability Density Function (PDF)
pdf_values = expon.pdf(x, scale=1/lambda_)
plt.plot(x, pdf_values, label='PDF', color='blue')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Exponential Distribution - PDF')
plt.legend()
plt.show()

# Cumulative Distribution Function (CDF)
cdf_values = expon.cdf(x, scale=1/lambda_)
plt.plot(x, cdf_values, label='CDF', color='green')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Exponential Distribution - CDF')
plt.legend()
plt.show()

# Survival Function
sf_values = expon.sf(x, scale=1/lambda_)
plt.plot(x, sf_values, label='Survival Function', color='red')
plt.xlabel('x')
plt.ylabel('Survival Probability')
plt.title('Exponential Distribution - Survival Function')
plt.legend()
plt.show()

# Hazard Function (PDF / Survival Function)
hazard_values = pdf_values / sf_values
plt.plot(x, hazard_values, label='Hazard Function', color='purple')
plt.xlabel('x')
plt.ylabel('Hazard Rate')
plt.title('Exponential Distribution - Hazard Function')
plt.legend()
plt.show()
```

```python
# Central Limit Theorem
from scipy.stats import norm

# Central Limit Theorem Demonstration
np.random.seed(42)  # Set seed for reproducibility

# Parameters
num_samples = 1000  # Number of samples per experiment
sample_size = 30  # Size of each sample
population_mean = 50  # True population mean
population_std = 15  # True population standard deviation

# Generate population data
population = np.random.normal(population_mean, population_std, 100000)

# Draw multiple samples and compute sample means
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Expectation (Mean)
expected_mean = np.mean(sample_means)
print(f"Expected Mean: {expected_mean}")

# Variance
variance = np.var(sample_means, ddof=1)  # Sample variance
print(f"Variance: {variance}")

# Standard Error
standard_error = np.std(sample_means, ddof=1)
print(f"Standard Error: {standard_error}")

# Z-Score Calculation (Example for a given sample mean)
sample_mean_example = 55  # Example sample mean
z_score = (sample_mean_example - expected_mean) / standard_error
print(f"Z-Score: {z_score}")

# Normal Approximation
x = np.linspace(min(sample_means), max(sample_means), 100)
pdf_values = norm.pdf(x, expected_mean, standard_error)
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='blue', label='Sample Means')
plt.plot(x, pdf_values, 'r-', label='Normal Approximation')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem - Normal Approximation')
plt.legend()
plt.show()

# Lyapunov CLT Condition Check
def lyapunov_condition(samples, delta=1.5):
    moments = [np.mean(np.abs(x - np.mean(x)) ** (2 + delta)) for x in samples]
    lhs = np.sum(moments) / (num_samples * (np.var(samples) ** (1 + delta / 2)))
    return lhs < np.inf

lyapunov_satisfied = lyapunov_condition(sample_means)
print(f"Lyapunov CLT Condition Satisfied: {lyapunov_satisfied}")

# Lindeberg CLT Condition Check
def lindeberg_condition(samples, epsilon=0.1):
    n = len(samples)
    total = np.sum([(x - np.mean(samples)) ** 2 * (np.abs(x - np.mean(samples)) > epsilon) for x in samples])
    return total / (n * np.var(samples)) < np.inf

lindeberg_satisfied = lindeberg_condition(sample_means)
print(f"Lindeberg CLT Condition Satisfied: {lindeberg_satisfied}")
```

```python
# Hypothesis Testing
# Basic Calculation Flow (4 Possibilities)
# Flow 1: 0,     one-Tailed Test  Compute Test Statistic  Compute p-value  Compare with   Decision
# Flow 2: H0, Ha    one-Tailed Test  Compute Test Statistic  Find Critical Value  Compare with Test Statistic  Decision
# Flow 3: H0, Ha    Two-Tailed Test  Compute Test Statistic  Compute p-value  Compare with   Decision
# Flow 4: H0, Ha    Two-Tailed Test  Compute Test Statistic  Find Critical Value  Compare with Test Statistic  Decision
```

```python
import scipy.stats as stats
import numpy as np

def hypothesis_test(test_type, stat_type, alpha, sample_stat, df=None):
    """
    Perform hypothesis testing using either the p-value method or critical value method.
    
    Parameters:
    - test_type: 'one-tailed' or 'two-tailed'
    - stat_type: 'z' for Z-test, 't' for T-test
    - alpha: Significance level
    - sample_stat: Test statistic (Z-score or t-score)
    - df: Degrees of freedom (only for T-test)
    
    Returns:
    - Decision: Reject or Fail to Reject H0
    """
    
    # Compute p-value based on test type
    if stat_type == 'z':
        if test_type == 'one-tailed':
            p_value = 1 - stats.norm.cdf(sample_stat)
        else:
            p_value = 2 * (1 - stats.norm.cdf(abs(sample_stat)))
    elif stat_type == 't' and df is not None:
        if test_type == 'one-tailed':
            p_value = 1 - stats.t.cdf(sample_stat, df)
        else:
            p_value = 2 * (1 - stats.t.cdf(abs(sample_stat), df))
    else:
        raise ValueError("Invalid statistic type or missing degrees of freedom for T-test.")
    
    # Compute critical value based on alpha level
    if stat_type == 'z':
        if test_type == 'one-tailed':
            critical_value = stats.norm.ppf(1 - alpha)
        else:
            critical_value = stats.norm.ppf(1 - alpha / 2)
    elif stat_type == 't' and df is not None:
        if test_type == 'one-tailed':
            critical_value = stats.t.ppf(1 - alpha, df)
        else:
            critical_value = stats.t.ppf(1 - alpha / 2, df)
    else:
        raise ValueError("Invalid statistic type or missing degrees of freedom for T-test.")
    
    # Decision making using p-value method
    decision_p = "Reject H0" if p_value < alpha else "Fail to Reject H0"
    
    # Decision making using critical value method
    decision_critical = "Reject H0" if abs(sample_stat) > critical_value else "Fail to Reject H0"
    
    return {
        "Test Type": test_type,
        "Statistic Type": stat_type,
        "Test Statistic": sample_stat,
        "P-Value": p_value,
        "Critical Value": critical_value,
        "Decision (P-Value Method)": decision_p,
        "Decision (Critical Value Method)": decision_critical
    }

# Example usage: One-tailed Z-test
alpha = 0.05
sample_z = 1.8  # Example test statistic for Z-test
result = hypothesis_test('one-tailed', 'z', alpha, sample_z)
for key, value in result.items():
    print(f"{key}: {value}")
```

```python
# Path 1: Confidence Interval Approach
# Flow 1 to 4
# Flow 1: H0, Ha    One-Tailed Test  Compute Test Statistic (Z)  Compute p-value  Compare with   Decision (Single Population Mean (Known Variance))
# Flow 2: H0, Ha    One-Tailed Test  Compute Test Statistic (t)  Compute p-value  Compare with   Decision (Single Population Mean (Unknown Variance))
# Flow 3: H0, Ha    Two-Tailed Test  Compute Test Statistic (Z)  Compute p-value  Compare with   Decision (Two Population Means (Independent Samples, Known Variances))
# Flow 4: H0, Ha    Two-Tailed Test  Compute Test Statistic (Pooled t)  Compute p-value  Compare with   Decision (Two Population Means (Independent Samples, Unknown but Equal Variances))
```

```python
import scipy.stats as stats

# Flow 1: Single Population Mean (Known Variance)
def single_mean_known_variance(sample_mean, population_mean, sigma, n, alpha, tail='one'):
    z = (sample_mean - population_mean) / (sigma / np.sqrt(n))  # Compute test statistic
    if tail == 'one':
        p_value = 1 - stats.norm.cdf(z)  # One-tailed test p-value
        critical_value = stats.norm.ppf(1 - alpha)
    else:
        p_value = 2 * (1 - stats.norm.cdf(abs(z)))  # Two-tailed test p-value
        critical_value = stats.norm.ppf(1 - alpha / 2)
    decision = 'Reject H0' if p_value < alpha else 'Fail to Reject H0'
    return z, p_value, critical_value, decision

# Flow 2: Single Population Mean (Unknown Variance)
def single_mean_unknown_variance(sample_mean, population_mean, sample_std, n, alpha, tail='one'):
    t = (sample_mean - population_mean) / (sample_std / np.sqrt(n))  # Compute test statistic
    df = n - 1  # Degrees of freedom
    if tail == 'one':
        p_value = 1 - stats.t.cdf(t, df)
        critical_value = stats.t.ppf(1 - alpha, df)
    else:
        p_value = 2 * (1 - stats.t.cdf(abs(t), df))
        critical_value = stats.t.ppf(1 - alpha / 2, df)
    decision = 'Reject H0' if p_value < alpha else 'Fail to Reject H0'
    return t, p_value, critical_value, decision

# Flow 3: Two Population Means (Independent Samples, Known Variances)
def two_means_known_variances(sample1_mean, sample2_mean, mu_diff, sigma1, sigma2, n1, n2, alpha, tail='one'):
    z = ((sample1_mean - sample2_mean) - mu_diff) / np.sqrt((sigma1**2 / n1) + (sigma2**2 / n2))
    if tail == 'one':
        p_value = 1 - stats.norm.cdf(z)
        critical_value = stats.norm.ppf(1 - alpha)
    else:
        p_value = 2 * (1 - stats.norm.cdf(abs(z)))
        critical_value = stats.norm.ppf(1 - alpha / 2)
    decision = 'Reject H0' if p_value < alpha else 'Fail to Reject H0'
    return z, p_value, critical_value, decision

# Flow 4: Two Population Means (Independent Samples, Unknown but Equal Variances)
def two_means_unknown_equal_variances(sample1_mean, sample2_mean, mu_diff, sample1_std, sample2_std, n1, n2, alpha, tail='one'):
    sp = np.sqrt(((n1 - 1) * sample1_std**2 + (n2 - 1) * sample2_std**2) / (n1 + n2 - 2))  # Pooled standard deviation
    t = ((sample1_mean - sample2_mean) - mu_diff) / (sp * np.sqrt(1/n1 + 1/n2))
    df = n1 + n2 - 2  # Degrees of freedom
    if tail == 'one':
        p_value = 1 - stats.t.cdf(t, df)
        critical_value = stats.t.ppf(1 - alpha, df)
    else:
        p_value = 2 * (1 - stats.t.cdf(abs(t), df))
        critical_value = stats.t.ppf(1 - alpha / 2, df)
    decision = 'Reject H0' if p_value < alpha else 'Fail to Reject H0'
    return t, p_value, critical_value, decision

# Example executions
print(single_mean_known_variance(105, 100, 15, 30, 0.05, 'one'))
print(single_mean_unknown_variance(105, 100, 15, 30, 0.05, 'one'))
print(two_means_known_variances(105, 100, 0, 15, 15, 30, 30, 0.05, 'one'))
print(two_means_unknown_equal_variances(105, 100, 0, 15, 15, 30, 30, 0.05, 'one'))
```

```python
# Flow 5: Compute Confidence Interval and Compare with Hypothesized Mean
def compute_confidence_interval(sample_mean, population_std, n, alpha):
    standard_error = population_std / np.sqrt(n)
    z_critical = stats.norm.ppf(1 - alpha / 2)
    margin_of_error = z_critical * standard_error
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error
    return ci_lower, ci_upper

# Flow 6: Compare Confidence Interval with Population Mean (H0)
def compare_confidence_interval(ci_lower, ci_upper, population_mean):
    return "Fail to reject H0" if ci_lower <= population_mean <= ci_upper else "Reject H0"

# Flow 7: One-tailed Test (Upper Tail) Using Confidence Interval
def one_tailed_upper_test(sample_mean, population_std, n, alpha, population_mean):
    standard_error = population_std / np.sqrt(n)
    z_critical = stats.norm.ppf(1 - alpha)
    ci_upper = sample_mean - z_critical * standard_error
    return "Fail to reject H0" if population_mean < ci_upper else "Reject H0"

# Flow 8: One-tailed Test (Lower Tail) Using Confidence Interval
def one_tailed_lower_test(sample_mean, population_std, n, alpha, population_mean):
    standard_error = population_std / np.sqrt(n)
    z_critical = stats.norm.ppf(1 - alpha)
    ci_lower = sample_mean + z_critical * standard_error
    return "Fail to reject H0" if population_mean > ci_lower else "Reject H0"

# Given Data
sample_mean = 5.2
population_std = 2.0
n = 30
alpha = 0.05
population_mean = 5.0

# Compute Results
ci_lower, ci_upper = compute_confidence_interval(sample_mean, population_std, n, alpha)
decision_flow6 = compare_confidence_interval(ci_lower, ci_upper, population_mean)
decision_flow7 = one_tailed_upper_test(sample_mean, population_std, n, alpha, population_mean)
decision_flow8 = one_tailed_lower_test(sample_mean, population_std, n, alpha, population_mean)

# Print Outputs
print((np.float64(ci_lower), np.float64(ci_upper)))
print(decision_flow6)
print(decision_flow7)
print(decision_flow8)
```

```python
# Path 2: Effect Size & Power Analysis Before Testing
# Flow 1: H0, Ha    Compute Effect Size (Cohen's d)  Compute Power  Compare with Desired Power  Decision
# Flow 2: H0, Ha    Compute Effect Size (Hedges' g)  Compute Power  Compare with Desired Power  Decision
# Flow 3: H0, Ha    Compute Effect Size (Glass' )  Compute Power  Compare with Desired Power  Decision
# Flow 4: H0, Ha    Compute Required Sample Size for a given Power  Compare with Available Sample Size  Decision
# Flow 5: H0, Ha    Compute Minimum Detectable Effect Size for a given Power  Compare with Expected Effect Size  Decision
# Flow 6: H0, Ha    Compute Effect Size  Compute Power  Determine if Sample Size Adjustments Are Needed  Decision
```

```python
import scipy.stats as stats
import statsmodels.stats.power as smp

# Effect Size & Power Analysis Before Testing (Flow 1 to 6)

# Given data
sample1 = np.array([5.1, 4.9, 5.2, 5.3, 5.0])  # Sample 1
sample2 = np.array([4.8, 4.7, 5.0, 5.1, 4.9])  # Sample 2
alpha = 0.05  # Significance level
desired_power = 0.8  # Desired power level

# Compute Effect Sizes
mean_diff = np.mean(sample1) - np.mean(sample2)
pool_std = np.sqrt(((np.std(sample1, ddof=1) ** 2) + (np.std(sample2, ddof=1) ** 2)) / 2)
cohen_d = mean_diff / pool_std  # Flow 1
hedges_g = cohen_d * (1 - (3 / (4 * (len(sample1) + len(sample2)) - 9)))  # Flow 2
glass_delta = mean_diff / np.std(sample2, ddof=1)  # Flow 3

print(f"Flow 1: Cohen's d = {cohen_d:.3f}")
print(f"Flow 2: Hedges' g = {hedges_g:.3f}")
print(f"Flow 3: Glass'  = {glass_delta:.3f}")

# Compute Power Analysis
power_analysis = smp.TTestIndPower()
power = power_analysis.power(effect_size=cohen_d, nobs1=len(sample1), alpha=alpha, ratio=1, alternative='two-sided')
print(f"Computed Power (Flow 1-3): {power:.3f}, Desired Power: {desired_power}")

# Flow 4: Compute Required Sample Size for Desired Power
desired_sample_size = power_analysis.solve_power(effect_size=cohen_d, alpha=alpha, power=desired_power, ratio=1)
print(f"Flow 4: Required Sample Size: {desired_sample_size:.1f}, Available Sample Size: {len(sample1)}")

# Flow 5: Compute Minimum Detectable Effect Size
min_effect_size = power_analysis.solve_power(nobs1=len(sample1), alpha=alpha, power=desired_power, ratio=1)
print(f"Flow 5: Minimum Detectable Effect Size: {min_effect_size:.3f}, Expected Effect Size: {cohen_d:.3f}")

# Flow 6: Decision Making Based on Effect Size & Power
if power >= desired_power:
    print("Flow 6: Power is sufficient, proceed with testing.")
else:
    print("Flow 6: Power is too low, consider increasing sample size or adjusting study design.")
```

```python
import scipy.stats as stats
from statsmodels.stats.multitest import multipletests

# Given data
np.random.seed(42)
data_group1 = np.random.normal(50, 10, 30)
data_group2 = np.random.normal(52, 10, 30)
data_group3 = np.random.normal(48, 10, 30)
data_group4 = np.random.normal(51, 10, 30)
data_group5 = np.random.normal(49, 10, 30)

groups = [data_group1, data_group2, data_group3, data_group4, data_group5]
num_tests = len(groups) - 1
alpha = 0.05

# Compute test statistics and p-values
p_values = []
for i in range(1, len(groups)):
    t_stat, p_val = stats.ttest_ind(groups[0], groups[i], equal_var=False)
    p_values.append(p_val)

# Flow 1: Bonferroni Correction
bonferroni_corrected = multipletests(p_values, alpha=alpha, method='bonferroni')
print("Flow 1 - Bonferroni Correction:", bonferroni_corrected[0])

# Flow 2: Holm-Bonferroni Correction
holm_corrected = multipletests(p_values, alpha=alpha, method='holm')
print("Flow 2 - Holm-Bonferroni Correction:", holm_corrected[0])

# Flow 3: Benjamini-Hochberg Correction
benjamini_corrected = multipletests(p_values, alpha=alpha, method='fdr_bh')
print("Flow 3 - Benjamini-Hochberg Correction:", benjamini_corrected[0])

# Flow 4: idk Correction
sidak_alpha = 1 - (1 - alpha) ** (1 / num_tests)
sidak_corrected = [p_val < sidak_alpha for p_val in p_values]
print("Flow 4 - idk Correction:", sidak_corrected)

# Flow 5: False Discovery Rate (FDR)
fdr_corrected = multipletests(p_values, alpha=alpha, method='fdr_tsbh')
print("Flow 5 - False Discovery Rate (FDR):", fdr_corrected[0])
```

```python
# Path 4: Bayesian Approach Instead of Frequentist Testing
# Flow 1: Define Prior Distributions  Likelihood Function  Compute Posterior Distribution  Compute Posterior Probability of 0 and a  Decision
# Flow 2: Define Prior Distributions  Compute Bayes Factor (BF)  Compare BF Threshold  Decision
# Flow 3: Define Prior Distributions  Compute Credible Interval (CI)  Check if 0 Value Falls Within CI  Decision
# Flow 4: Define Prior Distributions  Compute Expected Loss/Utility  Decision Based on Minimizing Loss or Maximizing Utility
```

```python
# Flow 1: Compute Posterior Probability of H0 and Ha
prior_h0 = 0.5  # Prior probability of H0
prior_ha = 1 - prior_h0
likelihood_h0 = stats.norm.pdf(5.2, loc=5.0, scale=0.5)  # Likelihood under H0
likelihood_ha = stats.norm.pdf(5.2, loc=5.5, scale=0.5)  # Likelihood under Ha
posterior_h0 = (likelihood_h0 * prior_h0) / (likelihood_h0 * prior_h0 + likelihood_ha * prior_ha)
posterior_ha = 1 - posterior_h0
print(f"Flow 1: Posterior Probabilities - P(H0)={posterior_h0:.4f}, P(Ha)={posterior_ha:.4f}")

# Flow 2: Compute Bayes Factor (BF)
bayes_factor = likelihood_h0 / likelihood_ha
print(f"Flow 2: Bayes Factor (BF) = {bayes_factor:.4f}")
if bayes_factor > 3:
    print("Evidence supports H0")
elif bayes_factor < 1/3:
    print("Evidence supports Ha")
else:
    print("Inconclusive evidence")

# Flow 3: Compute 95% Credible Interval (CI)
posterior_mean = 5.2  # Assumed posterior mean from Bayesian update
posterior_std = 0.5 / np.sqrt(30)  # Posterior standard deviation
credible_interval = stats.norm.interval(0.95, loc=posterior_mean, scale=posterior_std)
print(f"Flow 3: 95% Credible Interval: {credible_interval}")
if credible_interval[0] <= 5.0 <= credible_interval[1]:
    print("Fail to reject H0 based on credible interval")
else:
    print("Reject H0 based on credible interval")

# Flow 4: Compute Expected Loss/Utility
def expected_loss(decision, loss_h0=1, loss_ha=5):
    return decision * loss_h0 * posterior_h0 + (1 - decision) * loss_ha * posterior_ha

decision_h0 = expected_loss(1)
decision_ha = expected_loss(0)
print(f"Flow 4: Expected Loss if H0 is chosen: {decision_h0:.4f}, if Ha is chosen: {decision_ha:.4f}")
if decision_h0 < decision_ha:
    print("Decision: Favor H0")
else:
    print("Decision: Favor Ha")
```

```python
# Path 5: Sequential Testing Instead of Fixed Sample Testing
# Flow 1: H0, Ha    Likelihood Ratio  Compare with Decision Boundary  Decision
# Flow 2: H0, Ha    Compute Z-Score  Compare with Boundary at Each Stage  Decision
# Flow 3: H0, Ha    Continuous Monitoring  Compute p-value  Adjust for Multiple Looks  Decision
# Flow 4: H0, Ha    Pocock's Boundaries  Compare Test Statistic  Decision
# Flow 5: H0, Ha    O'Brien-Fleming's Boundaries  Compare Test Statistic  Decision
# Flow 6: H0, Ha    Alpha Spending Function (Lan-DeMets)  Compute Test Statistic  Decision
```

```python
# Given Parameters
alpha = 0.05  # Significance level
n = 30  # Sample size
observed_mean = 5.2  # Observed sample mean
population_mean = 5.0  # Hypothesized mean
std_dev = 2.0  # Population standard deviation

# Compute Standard Error
standard_error = std_dev / np.sqrt(n)

# Flow 1: Likelihood Ratio Test
z_score = (observed_mean - population_mean) / standard_error
likelihood_ratio = np.exp(-0.5 * z_score**2)
critical_value = np.exp(-0.5 * stats.norm.ppf(1 - alpha)**2)
print(f"Flow 1: Likelihood Ratio = {likelihood_ratio:.4f}, Decision: {'Reject H0' if likelihood_ratio < critical_value else 'Fail to Reject H0'}")

# Flow 2: Compute Z-Score and Compare with Decision Boundaries
z_critical = stats.norm.ppf(1 - alpha)
print(f"Flow 2: Z-Score = {z_score:.4f}, Decision: {'Reject H0' if z_score > z_critical else 'Fail to Reject H0'}")

# Flow 3: Continuous Monitoring - Adjusting p-value for Multiple Looks
p_value = 1 - stats.norm.cdf(z_score)
adjusted_alpha = alpha / 5  # Example: Adjusted for 5 looks
print(f"Flow 3: p-value = {p_value:.4f}, Decision: {'Reject H0' if p_value < adjusted_alpha else 'Fail to Reject H0'}")

# Flow 4: Pocock's Boundaries (Fixed boundary approach)
pocock_boundary = stats.norm.ppf(1 - alpha / np.log(n))
print(f"Flow 4: Pocock Boundary = {pocock_boundary:.4f}, Decision: {'Reject H0' if z_score > pocock_boundary else 'Fail to Reject H0'}")

# Flow 5: O'Brien-Fleming Boundaries (More conservative early on)
obrien_fleming_boundary = stats.norm.ppf(1 - alpha / (2 * np.log(n)))
print(f"Flow 5: O'Brien-Fleming Boundary = {obrien_fleming_boundary:.4f}, Decision: {'Reject H0' if z_score > obrien_fleming_boundary else 'Fail to Reject H0'}")

# Flow 6: Alpha Spending Function (Lan-DeMets Approach)
alpha_spent = alpha * (np.log(n) / np.log(n + 1))  # Example spending function
spending_boundary = stats.norm.ppf(1 - alpha_spent)
print(f"Flow 6: Alpha Spending Function Boundary = {spending_boundary:.4f}, Decision: {'Reject H0' if z_score > spending_boundary else 'Fail to Reject H0'}")
```

```python
from statsmodels.stats.multitest import multipletests

# Addressing P-Hacking & Multiple Testing Issues

# Sample p-values from multiple hypothesis tests
p_values = np.array([0.01, 0.03, 0.04, 0.005, 0.02, 0.06, 0.07, 0.001])
alpha = 0.05  # Significance level

# Flow 1: Multiple Testing Corrections (Bonferroni, Holm-Bonferroni)
bonferroni_corrected = multipletests(p_values, alpha=alpha, method='bonferroni')
holm_corrected = multipletests(p_values, alpha=alpha, method='holm')
print(f"Flow 1: Bonferroni Adjusted P-values: {bonferroni_corrected[1]}")
print(f"Flow 1: Holm Adjusted P-values: {holm_corrected[1]}")

# Flow 2: False Discovery Rate (FDR) Control (Benjamini-Hochberg, Benjamini-Yekutieli)
bh_corrected = multipletests(p_values, alpha=alpha, method='fdr_bh')
by_corrected = multipletests(p_values, alpha=alpha, method='fdr_by')
print(f"Flow 2: Benjamini-Hochberg Adjusted P-values: {bh_corrected[1]}")
print(f"Flow 2: Benjamini-Yekutieli Adjusted P-values: {by_corrected[1]}")

# Flow 3: Sequential Testing (Using Pocock's Method)
stagewise_alpha = alpha / np.sqrt(len(p_values))  # Pocock correction
significant_results = p_values < stagewise_alpha
print(f"Flow 3: Pocock Sequential Testing Results: {significant_results}")

# Flow 4: Bayesian Analysis Alternative (Computing Bayes Factor for each test)
def compute_bayes_factor(p):
    return (1 - p) / p  # Simplified Bayes Factor estimation

bayes_factors = np.array([compute_bayes_factor(p) for p in p_values])
threshold = 3  # Common threshold for strong evidence
bayesian_decision = bayes_factors > threshold
print(f"Flow 4: Bayes Factors: {bayes_factors}")
print(f"Flow 4: Bayesian Decision (Threshold {threshold}): {bayesian_decision}")

# Flow 5: Detecting & Addressing P-Hacking (Transparency in Reporting)
def check_p_hacking(p_vals, alpha=0.05):
    suspicious = np.sum(p_vals < alpha) / len(p_vals) > 0.5  # Arbitrary threshold
    return "Potential P-Hacking Detected" if suspicious else "No Evidence of P-Hacking"

p_hacking_check = check_p_hacking(p_values)
print(f"Flow 5: {p_hacking_check}")
```
